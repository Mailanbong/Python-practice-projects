```python
df = pd.DataFrame({'Courses': ["Python","Python","Python","pandas","Java"],
                    'Fee' : [20000,25000,30000,24000,40000],
                    'Duration':['30day','40days','60days','55days','50days']})
```

```python
df
```

```python
groupdf= pd.DataFrame(df.groupby('Courses').agg({'Courses':'count','Fee':'median','Duration':'min'}))
groupdf
```
